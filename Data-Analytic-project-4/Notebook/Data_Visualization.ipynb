```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Dataset\Cleaned_Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
product_counts = df["Product"].value_counts()

plt.figure(figsize=(8,5))
product_counts.plot(kind="bar")

plt.title("Product Distribution")
plt.xlabel("Product")
plt.ylabel("Number of Orders")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Bar_Chart.png")

plt.show()
```

```python
payment_counts = df["PaymentMethod"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(payment_counts,
        labels=payment_counts.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Payment Method Distribution")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Pie_Chart.png")

plt.show()
```

```python
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["TotalPrice"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Line_Chart.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["TotalPrice"], bins=10)

plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Histogram.png")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["TotalPrice"])

plt.title("Box Plot of Total Price")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Box_Plot.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Quantity"], df["TotalPrice"])

plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.savefig(r"C:\Users\osman\OneDrive\Desktop\Decode labs\Data-Analytic-project-4\Charts\Scatter_Plot.png")

plt.show()
```
